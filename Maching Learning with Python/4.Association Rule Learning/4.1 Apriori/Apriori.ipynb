### Importing the libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing the dataset

```python
df = pd.read_csv('Market_Basket_Optimisation.csv',header = None) #Don't make header
```

### Review data

```python
df.info()
```

```python
df.head()
```

### Get transactions from dataset

```python
transactions = [] #Create vector for storing transactions
```

```python
for i in range(0, np.shape(df)[0]):
    transactions.append([str(df.values[i,j]) for j in range(0,np.shape(df)[1])]) 
```

```python
np.shape(transactions)
```

### Training Apriori on the dataset  

```python
from apyori import apriori
rules = apriori(transactions, min_support= 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2 )
```

### Visualising the results


```python
results = list(rules) 
```

```python
results
```

```python
the_rules = []
for result in results:
    the_rules.append({'rule': ','.join(result.items), 'support':result.support,\
                      'confidence':result.ordered_statistics[0].confidence,\
                      'lift':result.ordered_statistics[0].lift})
df_rule = pd.DataFrame(the_rules, columns = ['rule', 'support', 'confidence', 'lift'])
```
